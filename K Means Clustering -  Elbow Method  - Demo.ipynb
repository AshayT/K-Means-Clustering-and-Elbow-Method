## K Means Clustering with Elbow method to find the optimal number of clusters

```python
#importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
#generating random:- sample size = 555, number of clusters = 5, standard deviation of clusters = 1
X, y = make_blobs(n_samples = 555, centers = 5, cluster_std = 0.9)
```

```python
#plotting the data to see the clusters
plt.figure(figsize=(10,7.5))
plt.scatter(X[:,0], X[:,1], s=10)
```

```python
#trying different values of k and saving the 'sum of squared distances' for each value of k 
sum_of_squared_distances = []
try_k = list(range(1,20))
for i in try_k:
    km = KMeans(n_clusters = i)
    km = km.fit(X)
    sum_of_squared_distances.append(km.inertia_)
```

#### Elbow method to find the number of clusters

```python
#visualizing the variation in 'sum of squared distances' w.r.t. value of k 
plt.plot(try_k, sum_of_squared_distances, 'bx--')
plt.title('Elbow method to find the optimal value of k')
plt.xlabel('Value of k')
plt.ylabel('Sum of squared distances')
```

#### The above graph clearly shows the optimal of value of k is 5

```python
#using the KMeans clustering algorithm to find the clusters
KM = KMeans(n_clusters = 5)
KM.fit(X)
clusters = KM.predict(X)
```

```python
#plotting the data to see the clusters
plt.figure(figsize=(10,7.5))
plt.scatter(X[:,0], X[:,1], c = clusters, cmap = 'viridis', s = 10)
```

#### Plotting the clusters with their centers

```python
#getting the coordinates of the centers of the clusters
centers = KM.cluster_centers_
```

```python
#plotting clusters with their centers
plt.figure(figsize=(10,7.5))
plt.scatter(X[:,0], X[:,1], c = clusters, cmap = 'viridis', s = 10)
plt.scatter( centers[:,0], centers[:,1], c = 'red', s = 50)
```
